# Chains

# First Chain

```python
# Task -1 [Prompt]

from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant"),
    ("human", "{input}"),
])
```

```python
# Task -2 [LLM]

from langchain_groq import ChatGroq

llm = ChatGroq(model="llama-3.1-8b-instant", temperature=0)
```

```python
# we need to always write .content to fetch the content 
# of the response from llm, so instead of that, we can write
# StringOutputParser

# Task -3 [Str Parser]

from langchain_core.output_parsers import StrOutputParser
str_parser = StrOutputParser()
```

### Manual Invocation

```python
template = prompt_template.invoke({"input": "What is the capital of France?"})
res = llm.invoke(template)
print(res.content)
```

### Chain Invocation

```python
# method -1

chain = prompt_template | llm | str_parser
res = chain.invoke({"input": "What is the capital of India?"})
print(res) # no need for .content
```

```python
# method 2 - using RunnableSequence 

from langchain_core.runnables import RunnableSequence

chain_1 = RunnableSequence(prompt_template, llm, str_parser)

chain_1.invoke({"input": "What is the capital of India?"})
```
