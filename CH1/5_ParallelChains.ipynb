```python
# Task -1
from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate([
    ("system", "You are a movie summarizer"),
    ("human", "please summarize the movie in 2-3 lines : {input}")
])
```

```python
# Task -2 
from langchain_groq import ChatGroq

llm = ChatGroq(model="llama-3.1-8b-instant",temperature=0)
```

```python
#  Task -3
from langchain_core.output_parsers import StrOutputParser

str_parser = StrOutputParser()
```

```python
# Task-4 [Custom Runnable]

from langchain_core.runnables import RunnableLambda

def dictionary_maker(text: str) -> dict:
    return {"text": text}

dictionary_maker_runnable = RunnableLambda(dictionary_maker) # this will
# make our normal function as a function to be passed in chain
```

## here we need to create parallel chains
and there are 2 options to create it -

1) to create a dedicated chain for chain-1 and it will do its own task

2) Easy - create a function for the entire chain-1 and chain-2

# Parallel Chain -1

```python
# task-1 [Prompt]

linkedIn_prompt = ChatPromptTemplate([
    ("system","you are a linkedin post generator"),
    ("human","Create a post for the following text for LinkedIn: {text}")
])

# Task -2 [LLM]
llm = ChatGroq(model="llama-3.1-8b-instant",temperature=0)

# Task -3 [Str Output Parser]

str_parser = StrOutputParser()

chain_linkedIn = linkedIn_prompt | llm | str_parser
```

# Parallel Chain 2

```python
from langchain_core.runnables import RunnableParallel, RunnableLambda
```

```python
def insta_chain(text: dict):
    text = text["text"]
    insta_prompt = ChatPromptTemplate([
    ("system","you are a Instagram post generator"),
    ("human","Create a post for the following text for Instagram: {text}")
    ])

    # Task -2 [LLM]
    llm = ChatGroq(model="llama-3.1-8b-instant",temperature=0)

    # Task -3 [Str Output Parser]
    str_parser = StrOutputParser()

    insta_chain = insta_prompt | llm | str_parser

    result = insta_chain.invoke(text)
    return result

insta_chain_runnable = RunnableLambda(insta_chain)
```

## Final Orchestration

```python
final_chain = (
    prompt_template | 
    llm |
    str_parser | 
    dictionary_maker_runnable |
    RunnableParallel(branches = {"linkedin" : chain_linkedIn, "instagram" : insta_chain_runnable})
)
```

```python
final_chain.invoke("Stuart Little")
```
