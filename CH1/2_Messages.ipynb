### Messages

```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from langchain_groq import ChatGroq

llm_groq = ChatGroq(model="llama-3.1-8b-instant", temperature=0)
my_messages = [
    SystemMessage(content="You are a gen-z teacher who teaches students with a lot of slang and emojis"),
    HumanMessage(content="What is Langchain")
]
```

```python
llm_groq.invoke(my_messages).content
```

### Prompts 

Prompts are the messages that are send to the LLM's

## Diff b/w prompts and messages ?

Messages - are static in nature and not user friendly

Prompts - are dynamic and more user friendly

```python
from langchain_core.prompts import PromptTemplate

user_input = input("What topic do you want a fun fact about? ")
dynamic_prompt = PromptTemplate.from_template("Write a fun fact about {topic}")

final_prompt = dynamic_prompt.invoke({"topic": user_input})

llm_groq.invoke(final_prompt).content
```

```python
from langchain_core.prompts import ChatPromptTemplate  #here we can set the tone of the conversation
# here we dont write SystemMessage / AIMessage / HumanMessage, we just write the content and the type of message will be decided by the ChatPromptTemplate

prompt_template = ChatPromptTemplate.from_messages(
    [
        ("system","you are a polite assistant"),
        ("user","write a fun fact about {topic}")
    ]
)

user_input = input("What topic do you want a fun fact about? ")
final_prompt = prompt_template.invoke({"topic": user_input})
llm_groq.invoke(final_prompt).content
```
