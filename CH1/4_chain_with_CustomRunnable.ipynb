```python
# Task -1 [Prompt]

from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant"),
    ("human", "{input}"),
])
```

```python
# Task -2 [LLM]

from langchain_groq import ChatGroq

llm = ChatGroq(model="llama-3.1-8b-instant", temperature=0)
```

```python
# we need to always write .content to fetch the content 
# of the response from llm, so instead of that, we can write
# StringOutputParser

# Task -3 [Str Parser]

from langchain_core.output_parsers import StrOutputParser
str_parser = StrOutputParser()
```

```python
# Task-4 [Custom Runnable]

from langchain_core.runnables import RunnableLambda

def dictionary_maker(text: str) -> dict:
    return {"text": text}

dictionary_maker_runnable = RunnableLambda(dictionary_maker) # this will
# make our normal function as a function to be passed in chain
```

```python
# Task - 5 [template for post]

prompt_post = ChatPromptTemplate.from_messages(
    messages=[
    ("system", "You are a social media post generator"),
    ("human", "Write a post for following text for linkedIn: {text}")
    ]
)
```

```python
# Task 6 [ generate post ]
llm = ChatGroq(model="llama-3.1-8b-instant", temperature=0)
```

```python
# task -7 [STR PARSER]

str_parser = StrOutputParser()
```

```python
# CHAIN

chain = prompt_template | llm | str_parser | dictionary_maker_runnable | prompt_post | llm | str_parser

chain.invoke("What is the capital of India?")
```
